# Borrower APR

```python
ticks =list(np.arange(0, 0.5, 0.025))
labels = ['{:.2f}'.format(v) for v in ticks]
bins = np.arange(0, df_Loan['BorrowerAPR'].max()+.02, .02)
plt.hist(data = df_Loan, x = 'BorrowerAPR', bins = bins)
plt.xticks(ticks, labels)
plt.xticks(rotation = 90)
plt.xlabel('Borrower APR')
plt.ylabel('Count')
plt.title("Borrower APR Count")
```

```python
df_Loan['BorrowerAPR'].describe()
```

The Borrower's Annual Percentage Rate (APR) for the loan.The above graph is a right-skewed distribution. Right-skewed distributions are also called positive-skew distributions. Thatâ€™s because there is a long tail in the positive direction on the number line. The mean is also to the right of the peak. A right-skewed distribution will have the mean to the right of the median.
This shows that the borrower rate is inversely propotional to the frequency of load as the borrower rate increses number of loans given decreases.
