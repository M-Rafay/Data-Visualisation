 ### 6b)  What is really of interest to us is what distinguishes completed loans from defaulted loans?
 
 Because there is no way to tell whether "current" loans will eventually default or not, we can't use them for our analysis. Nearly half of the dataset is "current loans" & is not useful to us, as the loans are still outstanding. In order to be conservative in our eventual estimates, to simplify the problem, and to retain data, let's assume all the "past due" and "chargedoff" loans (and that 1 cancellation) will default. Thus we'll be left with two classes: "completed" and "defaulted". Let's encode those binary outcomes as 1 and 0, respectively.

```python
#Remove Current/outstanding loans ( We have removed the Loan that are current because we cant say about them if they will be completed or will miss their Payment and defalut)

df_historical = df_0[df_0["LoanStatus"] != "Current"]  # dataset of only completed or defaulted values
df_historical
```

```python
df_historical["LoanStatus"].value_counts()
```

```python
#Encode all completed loans as 1, and all delinquent, chargedoff, cancelled and defaulted loans as 0

df_historical["LoanStatus"] = (df_historical["LoanStatus"] == "Completed").astype(int)

df_historical["LoanStatus"][:10]
```

```python
df_historical["LoanStatus"].describe()
```

```python
sns.distplot(df_historical["LoanStatus"])
plt.title('Distribution Plot')
```

```python
fig = plt.figure(figsize=(11.69, 8.27))

ax1 = fig.add_subplot(221)

sns.countplot(df_historical["LoanStatus"],palette="Set3")

ax2 = fig.add_subplot(222)
sns.barplot(y=df_historical["LoanStatus"]).set_ylim([0,1])
ax2.legend(title="Completed Loan %",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.xlabel('Completed Loan')
```

```python
print("Percentage Loan Completed: ",(df_historical["LoanStatus"].mean() *100),"%") 
print('\nPercentage of Loan Defaulted: ',((1 - df_historical["LoanStatus"].mean())*100),"%") 
```
