## converting our LoanStatus variable
into a binary variable excluding current loans, and visualize some of the relationships between default and other variables.

```python
df_original["LoanStatus"].value_counts()
```

The high number of loans have Current loan status.Now we are going to compare on the basis of completed or not completed

```python
#Remove current loans

df_Status_bool = df_original[df_original["LoanStatus"] != "Current"]

df_Status_bool["LoanStatus"].value_counts()
```

```python
#Encode all completed loans as 1, and all delinquent, chargedoff, cancelled and defaulted loans as 0

df_Status_bool["LoanStatus"] = (df_Status_bool["LoanStatus"] == "Completed").astype(int)

df_Status_bool["LoanStatus"][:10]
```

## Visualisation completed and not completed loans

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
sns.countplot(df_Status_bool["LoanStatus"])

ax2 = fig.add_subplot(222)
# sns.pie(df_Status_bool["LoanStatus"])
df_Status_bool["LoanStatus"].value_counts().plot.pie(figsize=(6, 6))
```

Number of completed status is more than the incompleted status


## Remember: 1 shows the completed  and 0 shows not completed

```python
df_Status_bool["LoanStatus"].mean(), 1 - df_Status_bool["LoanStatus"].mean()
```

So, in the historical data as a whole, 67.43% of loans are completed. 32.57% of loans "defaulted".

## now we will start exploring relationship between columns

```python
sns.barplot(x="BorrowerState", y="LoanStatus", data=df_Status_bool)
```

```python
df_Status_bool["BorrowerState"].value_counts()
```

Comparing two states:

```python
CA, FL = df_Status_bool[df_Status_bool["BorrowerState"] == "CA"], df_Status_bool[df_Status_bool["BorrowerState"] == "FL"]
CA["LoanStatus"].mean(), FL["LoanStatus"].mean()
```

the chances of getting loan in CA state is slightly higher than FL state

changing our numeric values to the actual category names, because the numeric values imply some sort of false ordinality, and we should really handle this like a categorical variable..

```python
df_Status_bool.replace(to_replace={"ListingCategory (numeric)": {0: "Unknown", 1: "Debt", 2: "Reno", 3: "Business", 4: "Personal",
                                                                5: "Student", 6: "Auto", 7: "Other", 8: "Baby", 9: "Boat", 
                                                                10: "Cosmetic", 11: "Engagement", 12: "Green", 13: "Household",
                                                                14: "LargePurchase", 15: "Medical", 16: "Motorcycle", 17: "RV",
                                                                18: "Taxes", 19: "Vacation", 20: "Wedding"}}, inplace=True)

df_Status_bool.rename(index=str, columns={"ListingCategory (numeric)": "ListingCategory"}, inplace=True)

df_Status_bool["ListingCategory"][:10]
```

```python
sns.barplot(x="ListingCategory", y="LoanStatus", data=df_Status_bool)
```

```python
Student, Personal = df_Status_bool[df_Status_bool["ListingCategory"] == "Student"], df_Status_bool[df_Status_bool["ListingCategory"] == "Personal"]

Student["LoanStatus"].mean(), Personal["LoanStatus"].mean()
```

By above analysis we can see that the loan chances of getting loan for students is greater than that of Personal
